  **_CRIAR UMA BASE DE DADOS:_**

    create database cadastro
    default character set utf8mb4
    default collate utf8mb4_0900_ai_ci;
    

  **_CRIAR UMA TABELA:_**

    create table if not exists pessoas(
    id int not null auto_increment,
    nome varchar(30) not null,
    nascimento date,
    sexo enum('M','F'),
    peso decimal(5,2),
    altura decimal(3,2),
    nacionalidade varchar(20) default 'Brasil',
    primary key (id)
    )default charset = utf8mb4;

**ALTERAR A ESTRUTURA DA TABELA:**
___

 **_Adicionar coluna_**
  
    alter table pessoas
    add column profissao varchar(10);

**_Adicionar coluna escolhendo sua POSIO_**

    alter table pessoas
    add column profissao varchar(10) after nome;
    
    alter table pessoas
    add column codigo int first;         <<- Adiciona codigo na 1 posio da coluna

**_Remover coluna_**

    alter table pessoas
    drop column profissao;

**_Remover tabela_**

    drop table if exists profissao;

**_Modificar definies_**

    alter table pessoas
    modify column profissao varchar (20);

**_Renomear coluna_** - _Mudar o nome da coluna de profissao para ocupacao._

    alter table pessoas
    change column profissao ocupacao varchar(20);
    

**_Renomear tabela_** - _Mudar o nome da Tabela

    alter table pessoas
    rename to humanos;

       1) Insero com todos os dados explcitos:_

    (id,nome,nascimento,sexo,peso,altura,nacionalidade)
    values
    ('1','Jessica Hemily','1994-03-09','F','60.50','1.66','Brasil');


    insert into pessoas

    values
    (default,'Jessica Hemily','1994-03-09','F','60.50','1.66','Brasil');

        2) No h necessidade de descrever id, devido o auto_increment!

    insert into pessoas
    (nome,nascimento,sexo,peso,altura,nacionalidade)
    values
    ('Jessica Hemily','1994-03-09','F','60.50','1.66','Brasil');
